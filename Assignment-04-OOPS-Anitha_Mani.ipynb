## Problem-1

Design a system for a library. Include classes for `Book`, `Patron`, and `Library`.

- The `Book` class should have attributes for title, author, ISBN, and a method `is_available()` that returns `True` if the book is not currently checked out and `False` otherwise. It should also have a method `check_out()` that marks the book as checked out and a method `check_in()` that marks it as available.
- The `Patron` class should have attributes for name and patron ID and a method `borrow_book(book)` that associates a book with the patron.
- The `Library` class should have a collection of `Book` objects and `Patron` objects. It should have methods to `add_book(book)`, `add_patron(patron)`, `lend_book(book, patron)`, and `return_book(book)`. The `lend_book` method should only allow a book to be lent if it's available and the patron exists in the library.


Test your implementation.

```python
class Book:
    def __init__(self,title,author,isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True
         
    def __str__(self):
         return (f"Title: {self.title}," f" Author: {self.author}, "f"ISBN: {self.isbn},"f" Available: {'Yes' if self.is_available else 'No'}")

    def check_in(self,book):
        if book.is_available:
            print(f"The Book '{book.title}' is available")
            return True
        else:   
            print(f"The Book '{book.title}' is not available")
            return False

    def check_out(self,book):
        if book.is_available:
            print(f"The Book '{book.title}' is not checked out/available") 
            return False
        else:
            print(f"The Book '{book.title}' is checked out") 
            return True

class Patron:
    def __init__(self,name,patronID):
        self.name=name
        self.patronID=patronID
        self.borrowed_books = []

    def __str__(self):
        return (f"Patron: {self.name}," f" PatronID: {self.patronID}," f"borrowed books: {[book.title for book in self.borrowed_books]}")
    
   
    def borrow_book(self,book):
        if book.is_available:
            self.borrowed_books.append(book)
            book.is_available = False
            return True
        else:
            return False
        
    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.is_available = True
            return True
        else:
            return False

class  Library:
    
    def __init__(self):
        self.books=[]
        self.patrons=[]

    def add_book(self,book):
        self.books.append(book)
        print(f"The Book '{book.title}' added to the library")

  
    def add_patron(self, patron):
        if patron.patronID not in self.patrons:
            self.patrons.append(patron)
            print(f"Patron {patron.name} added with ID {patron.patronID} to the library")
        else:
            print(f"Patron with ID {patron.patronID} already exists.")


    def lend_book(self,book,patron):
        if book in self.books and patron in self.patrons:
            if patron.borrow_book(book):
                print(f"{patron.name} has borrowed '{book.title}'")
            else:
                print(f"'{book.title}' is not available")
                
        else:
            print(f"Book or patron not found")
            
                
    def return_book(self,book,patron):
        if book in self.books and patron in self.patrons:
            if patron.return_book(book):
                print( f"{patron.name} has returned '{book.title}'")
            else:
                print( f"'{patron.name}' did not borrow '{book.title}'")
        else:
            print( "Book or patron not found")


book1= Book("Master Your Emotions","Thibaut Meurisse","21314324")
book2= Book("Harry Potter and the Philosophers Stone","J.K.Rowling","1232435")
book3= Book("The Compound Effect","Daren Hardy","75675756")


print(book1)
print(book2)
print(book3)

print()
patron1= Patron("Anitha","1")
patron2= Patron("Anu","2")
patron3= Patron("Akshatha","3")
 
print()

print(patron1)
print(patron2)
print(patron3)

print()

Library = Library()
Library.add_book(book1)
Library.add_book(book2)
Library.add_book(book3)
print()


Library.add_patron(patron1)
Library.add_patron(patron2)
Library.add_patron(patron3)

print("~~~~~~~~")

patron1.borrow_book(book1)
Library.lend_book(book2, patron1)
print(patron1)
Library.lend_book(book2, patron2)
Library.lend_book(book3, patron2)
Library.lend_book(book1, patron2)
Library.return_book(book2,patron1)
Library.lend_book(book3, patron3)
print(patron2)

book4 = Book("The Compound Effect2","Daren Hardy","75675752")
print(book4)
Library.add_book(book4)


```

## Problem-2

Create an base class `Shape` with an method `area()` and another method `perimeter()`. Then, create classes `Rectangle` and `Circle` that inherit from `Shape` and implement the `area()` method. The `perimeter()` method in `Shape` should raise a `NotImplementedError`. Implement the `perimeter()` method in `Rectangle` and `Circle`.

Test your implementation.

```python
class Shape:
    def __init__(self,name):
        self.name=name

    def area(self):
       pass

    def perimeter(self):
        raise NotImplementedError("Subclasses must implemet perimeter()")

class Rectangle(Shape):
    def __init__(self, name, length, breadth):
        super().__init__(name)
        self.length=length
        self.breadth=breadth
       
    def area(self):
        return self.length * self.breadth
    
    def perimeter(self):
        return (2 * (self.length + self.breadth))
    
class Circle(Shape):
    def __init__(self, name,radius):
        super().__init__(name)
        self.radius=radius
        
    
    def area(self):
       return 3.14 * self.radius**2
    
    def perimeter(self):
        return round(2 * 3.14 * self.radius)

rect = Rectangle("Rectangle",10,5)  

print(rect)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())

print()

circle = Circle("Circle",5)

print(circle)
print("Area:", circle.area())
print("Perimeter:", circle.perimeter())
```

```python
class Employee:
    def __init__(self,name,employee_id):
        self.name=name
        self.employee_id= employee_id

    def __str__(self):
        return (f"Name: {self.name}," f"Employee ID: {self.employee_id}")
    
class SalariedEmployee(Employee):
    def __init__(self, name, employee_id, perday_salary):
        super().__init__(name,employee_id)
        self.perday_salary = perday_salary

    def calculate_paycheck(self):
        return (30 * self.perday_salary)
    
class HourlyEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate,hours_worked):
        super().__init__(name,employee_id)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_paycheck(self):
        return (5 * (self.hourly_rate * self.hours_worked))

MonthlyEmp = SalariedEmployee("Anitha", "EMP001", 2500)
print(MonthlyEmp)
print("Monthly Paycheck:", MonthlyEmp.calculate_paycheck())

print()

HourlyEmp = HourlyEmployee("Akila","EMP002",500,6)
print(HourlyEmp)
print("Hourly Paycheck:",HourlyEmp.calculate_paycheck())
```

Design a system to model a simple online shopping cart. Create a class `Product` with attributes for `name` and `price`. Then, create a `ShoppingCart` class that has a list to store `Product` objects. Implement methods to `add_item(product)`, `remove_item(product_name)`, and `calculate_total()`.

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price= price

    def __str__(self):
        return (f"Name: {self.name}," f"Price: {self.price}")
    
class ShoppingCart:
    def __init__(self):
        self.items={}
        
    def add_item(self,product,quantity=1):
        #quantity=1
        if product in self.items:
            self.items[product] += quantity
        else:
            self.items[product] = quantity

    def remove_item(self,product,quantity=1):
        #quantity=1
        if product in self.items:
            self.items[product] -= quantity
            if self.items[product] <= 0:
                del self.items[product]

    def calculate_total(self):
        total=0
        for product, quantity in self.items.items():
            total += product.price * quantity
        print ("Total amount to be paid:",total)
        #return total
    
# Create  products
Carrot = Product("Carrot", 50)
WaterMelon = Product("WaterMelon", 80)
Potato = Product("Potato", 40)

print(Carrot)
print(WaterMelon)
print(Potato)
print()
# Create a shopping cart
cart = ShoppingCart()

# Add items to the cart
cart.add_item(Carrot, 2)
cart.add_item(WaterMelon, 1)
cart.add_item(Potato, 2)

cart.calculate_total()
print()

cart.remove_item(Carrot,1)
print("Updated amount post cart removal")
cart.calculate_total()
```
